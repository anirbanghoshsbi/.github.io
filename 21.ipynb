```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
```

```python
signals=pd.read_csv('/content/finally_final.csv',index_col=0) # comes from test file in aws.
```

```python
nifty=yf.download('^nsei',start='2021-04-25')
```

```python
signals.tail(2)
```

```python
signals.total.plot(kind='line',figsize=(15,5),color='r',marker='*');
```

```python
signals['final']=np.where(signals['total']>0,1,-1)
```

```python
signals.final.plot(kind='line',figsize=(15,5),color='r',marker='*');
```

```python
signals.reset_index(inplace=True)
nifty.reset_index(inplace=True)
```

```python
fig,ax=plt.subplots(2,1,figsize=(25,10),sharex=True)
ax[0].plot(signals.total,marker='*', label='signals')
ax[1].plot(nifty.Close,label='Close')
ax[0].axhline(y=0)
ax[0].axhline(y=5,color='g')
ax[0].axhline(y=-3,color='r')
plt.show()
```
