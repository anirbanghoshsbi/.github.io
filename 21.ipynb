```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
```

```python
signals=pd.read_csv('/content/nifty_signals.csv',index_col=0) # comes from test file in aws.
```

```python
nifty=yf.download('^nsei',start='2020-06-01')
```

```python
signals.tail(2)
```

```python
signals.total.plot(kind='line',figsize=(15,5),color='r');
```

```python
signals['final']=np.where(signals['total']>0,1,-1)
```

```python
signals.final.plot(kind='line',figsize=(15,5),color='r');
```

```python
fig,ax=plt.subplots(2,1,figsize=(25,10))
ax[0].plot(signals.total, label='signals')
ax[1].plot(nifty.Close,label='Close')
plt.show()
```
