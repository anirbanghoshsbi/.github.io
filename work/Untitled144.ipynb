```python
import yfinance as yf
import pandas as pd
import datetime
#from nsepython import *

# Install necessary packages if not already installed
# !pip install yfinance
# !pip install nsepython

# Get today's date and the date six months ago
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(days=6*30)
nifty200 = pd.read_csv('/content/nifty200_symbols.csv')

# List of Nifty50 stock symbols
nifty200_symbols =nifty200['Symbol']
# Initialize an empty list to store analysis results
analysis_results = []

for symbol in nifty200_symbols:
    # Fetch historical data
    data = yf.download(symbol, start=start_date.strftime('%Y-%m-%d'), end=end_date.strftime('%Y-%m-%d'))

    if data.empty:
        continue  # Skip if no data is returned

    # Calculate Return over the last six months
    start_price = data['Close'].iloc[0]
    end_price = data['Close'].iloc[-1]
    returns = (end_price - start_price) / start_price * 100

    # Fetch additional financial data
    ticker = yf.Ticker(symbol)
    info = ticker.info

    # Extract key financial metrics
    pe_ratio = info.get('trailingPE', None)
    pb_ratio = info.get('priceToBook', None)
    dividend_yield = info.get('dividendYield', 0) * 100 if info.get('dividendYield', None) else 0
    market_cap = info.get('marketCap', None)
    sector=info.get('sector')
    company_industry=info.get('industry')
    # Append the results
    analysis_results.append({
        'Symbol': symbol,
        '6M Return (%)': round(returns, 2),
        'P/E Ratio': pe_ratio,
        'P/B Ratio': pb_ratio,
        'Dividend Yield (%)': round(dividend_yield, 2),
        'Market Cap': market_cap,
        'sector':sector,
        'company_industry':company_industry
    })

# Convert the results into a DataFrame
df_results = pd.DataFrame(analysis_results)

# Sort by 6M Return
df_sorted = df_results.sort_values(by='6M Return (%)', ascending=False)


# Save the results to a CSV file
df_sorted.to_csv('nifty200_analysis.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL for Nifty 200 constituents
url = "https://www.nseindia.com/api/equity-stockIndices?index=NIFTY%20200"

# Headers to mimic a browser visit (NSE blocks scripts without headers)
headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate",
    "Referer": "https://www.nseindia.com/market-data/live-equity-market"
}

# Create a session
session = requests.Session()
session.headers.update(headers)

# Get cookies
session.get("https://www.nseindia.com", timeout=5)

# Fetch data
response = session.get(url, timeout=5)
data = response.json()

# Extract symbols and format them
symbols = [item['symbol'] + '.NS' for item in data['data']]

# Display the list
print(symbols)

# Optionally, save to a CSV file
df = pd.DataFrame(symbols, columns=['Symbol'])
df.to_csv('nifty200_symbols.csv', index=False)
```
