```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/anirbanghoshsbi/data/main/pcr.csv',sep=',')
```

```python
data.tail(2)
```

```python
data['Date']=pd.to_datetime(data['Date'],format='%Y-%m-%d')
```

```python
data=data.set_index('Date')
```

```python
fig,ax=plt.subplots(3,1,figsize=(25,10),sharex=True)
ax[0].plot(data['Nifty'], label='Nifty')
ax[1].plot(data['PCR'],color='g',label='Put Call Ratio')
ax[2].plot(data['WPCR'],color='r', label= 'Weighted put call ratio')
ax[0].legend()
ax[1].legend()
ax[2].legend()
plt.show()
```

```python
fig,ax=plt.subplots(2,1,figsize=(25,5),sharex=True)
ax[0].plot(data['Nifty'], label='Nifty')
ax[1].plot(data['PCR'].rolling(window=3).mean(),color='g',label='Put Call Ratio 10 EMA')
ax[1].plot(data['WPCR'].rolling(window=10).mean(),color='r', label= 'Weighted put call ratio 10 EMA')
ax[0].legend(loc='upper left')
ax[1].legend(loc='upper left')
plt.show()
```
