```python
import json
import  urllib
```

```python
instrument_url = "https://margincalculator.angelbroking.com/OpenAPI_File/files/OpenAPIScripMaster.json"
response = urllib.request.urlopen(instrument_url)
instrument_list = json.loads(response.read())
```

```python
# Extract unique names
unique_names = set(entry['lotsize'] for entry in instrument_list )#if 'NIFTY' in entry['name'].upper()
```

```python
unique_names
```

```python

# Conditions for filtering
desired_conditions = {'lotsize': '15',"name": "BANKNIFTY" ,"exch_seg": "NFO",'expiry':'29FEB2024'}

# Extracting entries based on conditions
filtered_data = [entry for entry in instrument_list if all(entry.get(key) == value for key, value in desired_conditions.items())]

# Display the result
for entry in filtered_data:
    print(entry)
```

```python
unique_names = set(entry['token'] for entry in filtered_data )
```
