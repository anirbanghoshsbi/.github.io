```python
import pandas as pd
import yfinance as yf

def get_nifty50_tickers():
    data = pd.read_csv('https://raw.githubusercontent.com/anirbanghoshsbi/data/main/STOCKS100.txt', sep='\t')
    nifty100 = data['yahoo finance ticker'].tolist()
    return nifty100


def fetch_stock_data(symbol, start_date, end_date):
    stock_data = yf.download(symbol, start=start_date, end=end_date)
    return stock_data

def calculate_ema(data, ema_period):
    return data['Close'].ewm(span=ema_period, adjust=False).mean()

start_date = '2023-10-01'
end_date = '2024-01-28'
ema_21_period = 5
ema_200_period = 35
top_decile_percentage = 0.1

nifty50_tickers = get_nifty50_tickers()
top_decile_stocks = []

for symbol in nifty50_tickers:
    stock_data = fetch_stock_data(symbol, start_date, end_date)

    if len(stock_data) > ema_200_period:
        ema_21 = calculate_ema(stock_data, ema_21_period)
        ema_200 = calculate_ema(stock_data, ema_200_period)

        distance = ema_21 - ema_200

        # Append relevant information to the list
        top_decile_stocks.append({
            'Symbol': symbol,
            'Close': stock_data['Close'].iloc[-1],
            'Distance': distance.iloc[-1],
        })

# Create DataFrame from the list and sort by 'Distance'
top_decile_df = pd.DataFrame(top_decile_stocks).sort_values(by='Distance', ascending=False)

# Calculate the index for the top decile
top_decile_index = int(len(top_decile_df) * top_decile_percentage)

# Extract the top decile stocks
top_decile_stocks = top_decile_df.head(top_decile_index)

print("Top Decile Stocks:")
print(top_decile_stocks)
```
