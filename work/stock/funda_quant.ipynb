```python
import pandas as pd

# Load the data
data = pd.read_csv('nifty-500.csv')

# Function to calculate composite score
def calculate_score(row):
    roe_score = row['Return on equity'] * 0.35  # 35% weight
    pe_score = (100 / row['Price to Earning']) * 0.10  # 10% weight, inverted as lower is better
    ps_score = (100 / row['Price to Sales']) * 0.10  # 10% weight, inverted as lower is better
    debt_equity_score = (1 / (1 + row['Debt to equity'])) * 0.2  # 20% weight, inverted as lower is better
    growth_score = row['Sales growth'] * 0.15  # 15% weight

    return roe_score + pe_score+ ps_score + debt_equity_score + growth_score
# Filter for market cap > 30000 crores
# Filter for market cap > 30000 crores and exclude banks
data_filtered = data[(data['Market Capitalization'] > 30000) & (~data['Industry'].str.contains('Banks', case=False))]
# Calculate scores
data_filtered['Score'] = data_filtered.apply(calculate_score, axis=1)
# Sort by score and select top 10
top_10 = data_filtered.sort_values('Score', ascending=False).head(50)
```

```python

# Group by industry and sort within each group
grouped = data_filtered.groupby('Industry').apply(lambda x: x.sort_values('Score', ascending=False))

# Function to get top companies in each industry
def get_top_companies(group):
    return group.head(50)  # Get top 2 companies from each industry

# Apply the function to each group
top_companies = grouped.apply(get_top_companies)

# Reset index for easier viewing
top_companies = top_companies.reset_index(drop=True)

# Display results
print(top_companies[['Industry', 'Name','Score']])
```
