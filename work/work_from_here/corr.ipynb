```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Fetch Nifty 50 index data
nifty_data = yf.download('^NSEI', start='2020-01-01', end='2023-04-07')

# Fetch Gold Bees ETF data
goldbees_data = yf.download('GOLDBEES.NS', start='2020-01-01', end='2023-04-07')

# Calculate daily returns
nifty_returns = nifty_data['Close'].pct_change().dropna()
goldbees_returns = goldbees_data['Close'].pct_change().dropna()

# Align the dates of the two datasets
aligned_data = pd.concat([nifty_returns, goldbees_returns], axis=1).dropna()
aligned_data.columns = ['Nifty', 'GoldBees']

# Calculate correlation
correlation = aligned_data.corr().iloc[0, 1]

# Plot the correlation graph
plt.scatter(aligned_data['Nifty'], aligned_data['GoldBees'], alpha=0.5)
plt.xlabel('Nifty Daily Returns')
plt.ylabel('Gold Bees Daily Returns')
plt.title(f'Correlation between Nifty 50 and Gold Bees ETF (Corr: {correlation:.2f})')
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Fetch Nifty 50 index data
nifty_data = yf.download('^NSEI', start='2020-01-01', end='2023-04-07')

# Fetch Gold Bees ETF data
goldbees_data = yf.download('GOLDBEES.NS', start='2020-01-01', end='2023-04-07')

# Calculate daily returns
nifty_returns = nifty_data['Close'].pct_change().dropna()
goldbees_returns = goldbees_data['Close'].pct_change().dropna()

# Align the dates of the two datasets
aligned_data = pd.concat([nifty_returns, goldbees_returns], axis=1).dropna()
aligned_data.columns = ['Nifty', 'GoldBees']

# Calculate rolling 60-day correlation
rolling_correlation = aligned_data['Nifty'].rolling(window=20).corr(aligned_data['GoldBees'])

# Create subplots
fig, ax = plt.subplots(2, 1, figsize=(10,3), sharex=True)

# Plot Nifty 50 Close Prices
ax[0].plot(nifty_data.index, nifty_data['Close'], label='Nifty 50 Close')
ax[0].set_title('Nifty 50 Close Prices')
ax[0].set_ylabel('Close Price')
ax[0].grid(True)

# Plot 60-day Rolling Correlation
ax[1].plot(rolling_correlation.index, rolling_correlation, label='25-Day Rolling Correlation', color='green')
ax[1].set_title('20-Day Rolling Correlation between Nifty 50 and Gold Bees ETF')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Rolling Correlation')
ax[1].grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
