```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
data = yf.download('^nsei',start='2019-01-05')
```

```python
data1 = yf.download('goldbees.bo',start='2019-01-05')
```

```python
data['corr_running']=data['Close'].rolling(window=14).corr(data1['Close'])
```

```python
data['corr_running'].fillna(data.corr_running.mean(),inplace=True)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2,1, figsize=(10,5),sharex=True)
ax[0].plot(data['corr_running'][-800:])
ax[1].plot(data['Close'][-800:])
ax[0].axhline(y=0)
plt.show()
```
