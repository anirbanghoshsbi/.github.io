```python
import pandas as pd
import numpy as np
```

```python

data=pd.read_csv('/content/rf_TA.csv',index_col='Date')
```

```python
data['signal']=data['pred0']+(data['pred1']+data['pred2'])/2
```

```python
data['signal']=np.where(data['signal']>0,1,-1)
```
